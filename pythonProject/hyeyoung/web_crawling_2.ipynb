```python
import re

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
from selenium.webdriver.common.keys import Keys


#브라우저 꺼짐 방지
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)

#불필요한 에러 메시지 없애기
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"]) #셀레니움 로그 무시

#서비스랑 드라이버 객체 생성. 세팅은 아니지만, 이거 안하면 어차피 진행이 안되서.
service = Service(executable_path=ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)


driver.get('https://instagram.com')
time.sleep(3)

#인스타 로그인
driver.implicitly_wait(10)
login_id = driver.find_element(By.CSS_SELECTOR, 'input[name="username"]')
login_id.send_keys('kimhy1867@naver.com') #아이디
login_pwd = driver.find_element(By.CSS_SELECTOR, 'input[name="password"]')
login_pwd.send_keys('kimhy6407$') #비번
driver.implicitly_wait(10)
login_pwd.submit()
time.sleep(2)
login_id.send_keys(Keys.ENTER)
button = driver.find_element(By.XPATH, '//button[text()="정보 저장"]')
button.click()
time.sleep(10)
```

```python
# 인스타 검색 결과 단계별로 들고 오기
## --- 인스타 검색 결과 페이지
## word를 입력해서 해쉬태그 검색하는 함수
def insta_searching(user_id):
    url = 'https://www.instagram.com/' + user_id
    return url

## 첫번째 게시글 열기
def select_first(driver):
    first = driver.find_element(By.CSS_SELECTOR,'div._aagu')
    first.click()
    time.sleep(5)

## 콘텐츠 들고 오기
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    try:
        content = soup.select('div._a9zr')[0].text
    except:
        content = ' '
    userid = ''
    try:
        userid_element = soup.select('h2')
        if userid_element:
            userid = userid_element[0].text.strip()
        else:
            print("No h2 element found.")
    except Exception as e:
        print(f"Error extracting h2 title: {e}")
    tags = re.findall(r'#[^\s#,\\]+', content)
    date = soup.select('time._a9ze._a9zf')[0]['datetime'][:10]
    data = [userid, content, date, tags]
    return data
```

```python
## 다음으로 넘기기
def move_next(driver):
    right = driver.find_element(By.CSS_SELECTOR, 'div._aaqg._aaqh')
    right.click()
    time.sleep(3)
```

```python
# list 내 user 순회
results = []
userids = ['s2isoda', 'gung_di_food','yumjji_','meok_tam']

for userid in userids:
    url = insta_searching(userid)
    driver.get(url)
    time.sleep(7)
    select_first(driver)  
    
    # 첫 게시글 열기

    
    target = 10
    for i in range(target):
        try:
            data = get_content(driver)
            results.append(data)
            move_next(driver)
        except:
            time.sleep(5)
            move_next(driver)
    print(results)
```
